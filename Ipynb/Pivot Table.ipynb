```python
import pandas as pd
import numpy as np
```

```python
combine_book_rating = pd.read_csv('../Dataset/main_data.csv')
books = pd.read_csv('../Dataset/books_change.csv')
```

```python
combine_book_rating['original_title'] = combine_book_rating['original_title'].str.title()
books['original_title'] = books['original_title'].str.title()
books = books.loc[:,['id','book_id','isbn','authors','original_publication_year','original_title','average_rating','ratings_count','image_url']]
```

```python
combine_book_rating
```

```python
combine_book_rating['original_title'] = combine_book_rating['original_title'].str.strip()
```

```python
final_dataset = combine_book_rating.pivot_table(index='user_id',columns='original_title',values='rating')
print(final_dataset.shape)
final_dataset
```

```python
final_dataset = final_dataset.fillna(0)
```

```python
final_dataset.to_csv('../Dataset/pivot.csv',index=False)
```

```python
final_dataset
```

```python
def recommendation(title):
    find_books = final_dataset[title]
    corr_books = final_dataset.corrwith(find_books, method='pearson')
    corr_table = pd.DataFrame(corr_books, columns=['correlation'])
    corr_table.sort_values('correlation', ascending = False)
    corr_table.dropna(inplace=True)
    corr_summary = corr_table.join(books['ratings_count'])
    corr_summary = pd.merge(corr_table, books, on='original_title')
    corr_summary = corr_summary.sort_values('correlation', ascending = False)
#     corr_summary = corr_summary[corr_summary['ratings_count']>=50000].sort_values('correlation', ascending = False)
    return corr_summary
```

```python
recommendation('The Lost World')
```

```python
find_books = final_dataset['A Great And Terrible Beauty']
corr_books = final_dataset.corrwith(find_books, method='pearson')
corr_table = pd.DataFrame(corr_books, columns=['correlation'])
corr_table.sort_values('correlation', ascending = False).head(10)
corr_table.dropna(inplace=True)
corr_summary = corr_table.join(books['ratings_count'])
corr_summary = pd.merge(corr_table, books, on='original_title')
corr_summary[corr_summary['ratings_count']>=1000000].sort_values('correlation', ascending = False)
# corr_summary[corr_summary['rating_count']>=1].sort_values('correlation', ascending = False)
# corr_summary.sort_values(['ratings_count','correlation'], ascending=[False, True])
```
